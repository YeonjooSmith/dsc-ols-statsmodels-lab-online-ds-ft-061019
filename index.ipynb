## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# import libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols
# build the formula 
f = 'sales ~ TV'
# create a fitted model in one line
model = ols(formula = f, data = df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
X_new = pd.DataFrame({'TV':[df.TV.min(),df.TV.max()]})
print(X_new.head())
# make predictions for those x values and store them
preds = model.predict(X_new)
print(preds)

# first, plot the observed data and the least squares line

df.plot(kind='scatter', x= 'TV',y= 'sales')#Sect4,Data Vis. with Pandas
plt.plot(X_new, preds, color ='r')
plt.show()
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig =plt.figure(figsize= (15,8))
fig =sm.graphics.plot_regress_exog(model,"TV", fig=fig)
plt.show()
```

### Record Your observations on heteroscedasticity
From first two plots, the variance is creating a cone-shape which is a sign of heteroscedasticity. (The residuals are heteroscedastic.) This break the assumption for using linear regression model. 

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
#line = '45' : 45 deg line, fit = True : the parameters for dist are fit automatically using dist.fit
fig.show()
```

#### statsmodels.graphics.gofplots.qqplot
https://www.statsmodels.org/devel/generated/statsmodels.graphics.gofplots.qqplot.html

### Record Your observations on the normality assumption
The residual is normally distributed which fits for the assumption.

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here
f2 = 'sales ~ radio'
model = ols(formula= f2, data = df).fit()
```

```python
model.summary()
```

```python
# create a DataFrame with the minimum and maximum values of TV
X_new = pd.DataFrame({'radio':[df.radio.min(),df.radio.max()]})
print(X_new.head())
# make predictions for those x values and store them
preds = model.predict(X_new)
print(preds)

# first, plot the observed data and the least squares line

df.plot(kind='scatter', x= 'radio',y= 'sales')#Sect4,Data Vis. with Pandas
plt.plot(X_new, preds, color ='r')
plt.show()
```

```python
fig =plt.figure(figsize= (15,8))
fig =sm.graphics.plot_regress_exog(model,"radio", fig=fig)
plt.show()
```

```python
# Code for QQ-plot here
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
#line = '45' : 45 deg line, fit = True : the parameters for dist are fit automatically using dist.fit
fig.show()
```

### Record your observations here for goodnes of fit

* $R^2$ for radio is only about a half of one for TV. So TV explains sales better than radio does.
* Kurtosis is higher than TV's which means that its peak is curvier than TV's. (?)
* The residuals QQ plot looks off, the normality assumption is not fulfilled. 
* There is obvious heteroscedasticity as with the case of TV. 
* The slope of the linear regression presents that one unit increase in radio advertisement spending is associated with a 0.2025 unit increase in Sales. This is more than the association between TV and Sales(0.0475).
